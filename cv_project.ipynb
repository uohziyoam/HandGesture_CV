```python
import numpy as np
import cv2
from pylab import *
from collections import deque
from wxpy import *
bot = Bot()

def _bodyskin_detetc(frame):
    ycrcb = cv2.cvtColor(frame, cv2.COLOR_BGR2YCrCb)
    (_, cr, _) = cv2.split(ycrcb)
    cr1 = cv2.GaussianBlur(cr, (5, 5), 0)
    _, skin = cv2.threshold(cr1, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
    return skin

def template_image(target,tpl,dst,whattodisplay):   
    min_val = 1
    index = 0
    for i in range(len(tpl)):
        result = cv2.matchTemplate(target, tpl[i] , cv2.TM_SQDIFF_NORMED)
        if min_val > cv2.minMaxLoc(result)[0]:
            min_val = cv2.minMaxLoc(result)[0]
            index = i
            
#         min_val = min(cv2.minMaxLoc(result)[0],min_val)
    if min_val < 0.45:
        dst=cv2.putText(dst,whattodisplay[index],(50,310),cv2.FONT_HERSHEY_SIMPLEX,5,(255,255,255),2)
#         my_friend = bot.friends().search('杰西', sex=FEMALE)[0]
#         my_friend.send('有人在向你挥手!')
    return target, dst

def rescale(image,scale):
    shape = image.shape
    dim = (image.shape[1]//scale,image.shape[0]//scale)
    resized = cv2.resize(image, dim)
    return resized

def findArea(src):
    area = 0
    for i in range(src.shape[0]):
        for j in range(src.shape[1]):
            if src[i,j] == 255:
                area +=1
    return area

def getContours(img):
    kernel = np.ones((5,5),np.uint8)
    closed = cv2.morphologyEx(img,cv2.MORPH_OPEN,kernel)
    closed = cv2.morphologyEx(closed,cv2.MORPH_CLOSE,kernel)

    contours,h = cv2.findContours(closed,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
    
    vaildContours = []
    for cont in contours:
        M = cv2.moments(cont)
        #get moments of each point
        #and calculate the center point according to the formula of calculating centroid
        cX = int(M["m10"] / M["m00"])
        cY = int(M["m01"] / M["m00"])
        
        if cv2.contourArea(cont)>9000:
            vaildContours.append(cv2.convexHull(cont))
    return  vaildContours,cX,cY

def framedifference(current_frame,previous_frame,dst):
    pixel = 0
    
    current_frame_gray = _bodyskin_detetc(current_frame)
    previous_frame_gray = _bodyskin_detetc(previous_frame)
        
    frame_diff = cv2.absdiff(current_frame_gray,previous_frame_gray)

    
    
    q.pop()
    q.append(frame_diff)
    dst = np.zeros((current_frame.shape[0],current_frame.shape[1],1), np.uint8)
    for i in range(dst.shape[0]):
        for j in range(dst.shape[1]):
            if q[0][i][j] == 255 or q[1][i][j] == 255 or q[2][i][j] == 255:
                dst[i][j] = 255 
#                 pixel += 1
    if (pixel >= current_frame.shape[0] * current_frame.shape[1] * 0.5):
        print("success")
        my_friend = bot.friends().search('杰西', sex=FEMALE)[0]
        my_friend.send('有人在向你挥手!')
    return dst

def init_queue(frame):
    height = frame.shape[0]
    width = frame.shape[1]
    init1 = np.zeros((height,width,1), np.uint8)
    init2 = np.zeros((height,width,1), np.uint8)
    init3 = np.zeros((height,width,1), np.uint8)
    q = deque()
    q.append(init1)
    q.append(init2)
    q.append(init3)
    return q

def init_template(tpl,scale):
    for i in range(len(tpl)):
        tpl[i] = _bodyskin_detetc(tpl[i])
        tpl[i] = rescale(tpl[i],scale) 
```

```python
import seaborn as sn
import pandas as pd
import matplotlib.pyplot as plt
array = [[12,4,4,0],[0,15,2,0],[0,1,14,0],[8,0,0,20]]
df_cm = pd.DataFrame(array, index = [i for i in "ABCD"],
                  columns = [i for i in "ABCD"])
plt.figure(figsize = (10,7))
sn.heatmap(df_cm, annot=True)
```

```python
import numpy as np

def precision(label, confusion_matrix):
    col = confusion_matrix[:, label]
    return confusion_matrix[label, label] / col.sum()
    
def recall(label, confusion_matrix):
    row = confusion_matrix[label, :]
    return confusion_matrix[label, label] / row.sum()
def precision_macro_average(confusion_matrix):
    rows, columns = confusion_matrix.shape
    sum_of_precisions = 0
    for label in range(rows):
        sum_of_precisions += precision(label, confusion_matrix)
    return sum_of_precisions / rows
def recall_macro_average(confusion_matrix):
    rows, columns = confusion_matrix.shape
    sum_of_recalls = 0
    for label in range(columns):
        sum_of_recalls += recall(label, confusion_matrix)
    return sum_of_recalls / columns

def accuracy(confusion_matrix):
    diagonal_sum = confusion_matrix.trace()
    sum_of_all_elements = confusion_matrix.sum()
    return diagonal_sum / sum_of_all_elements 

cm = np.array([[12,4,4,0],[0,15,2,0],[0,1,14,0],[8,0,0,20]])
cm = cm.transpose()
print("label precision recall")
for label in range(4):
    print(f"{label:5d} {precision(label, cm):9.3f} {recall(label, cm):6.3f}")
    
print("precision total:", precision_macro_average(cm))
print("recall total:", recall_macro_average(cm))
print("accuracy:", accuracy(cm))
```
